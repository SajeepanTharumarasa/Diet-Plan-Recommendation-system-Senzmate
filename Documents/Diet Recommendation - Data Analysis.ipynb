```python
#pattern1 = r'^(Weight (?:Gain|Loss))\s+([A-Za-z\s]+)?(?:None)?\s+([A-Za-z\s]+)?\s*([A-Za-z\s]+(?:Non Veg|Veg))\s*-\s*(\d+)(?: \(\d+\))?$'
pattern = r'(Weight (?:Gain|Loss))\s+([A-Za-z\s]+)?(?:None)?\s+([A-Za-z\s]+)?\s*([A-Za-z\s]+(?:Non Veg|Veg))\s*-\s*(\d+)'

def extract_parts(value):
    match = re.match(pattern, value)
    if match:
        return match.groups()
    else:
        return (None, None, None, None, None)

# Apply the function to the column and create new columns
dietPlan[['Type', 'Region', 'Condition', 'Diet', 'Value']] = dietPlan['details'].apply(extract_parts).apply(pd.Series)
dietPlan.head(20)
```
